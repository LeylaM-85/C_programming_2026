# C Proqramlaşdırma: Dəyişənlər və Yaddaş
Bu dərsimizdə biz müxtəlif məlumat tiplərinin RAM-da nə qədər yer tutduğunu öyrənəcəyik.

```python
# C kodunu notebook daxilində işlətmək üçün kiçik bir hiylə
c_kodu = """
#include <stdio.h>

int main() {
    printf("int tipi yaddasda %zu bayt yer tutur.\n", sizeof(int));
    printf("char tipi yaddasda %zu bayt yer tutur.\n", sizeof(char));
    printf("double tipi yaddasda %zu bayt yer tutur.\n", sizeof(double));
    return 0;
}
"""

with open("test.c", "w") as f:
    f.write(c_kodu)

!gcc test.c -o test
!./test
```
